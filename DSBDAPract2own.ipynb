```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = {
    'ID':[1,2,3,4,5],
    'Name':['Atharva', 'Kunal', 'Swayam', 'Tejasvi', 'Prem'],
    'Math':[85, 90, np.nan, 66, 125],
    'Science':[78, 88, 92, None, 65],
    'English':[80, 85, 70, 60, 75],
    'Attendance':[0.92, 0.85, 0.95, 0.40, 1.10]
}
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.loc[len(df)] = df.loc[1]
```

```python
df
```

```python
df[df.duplicated(subset=['Math'])]
```

```python
df = df.drop_duplicates().copy()
```

```python
df
```

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
```

```python
df
```

```python
df['Science'] = df['Science'].fillna(df['Science'].median())
```

```python
df
```

```python
def attend(x):
    return min(x, 1)

df['Attendance'] = df['Attendance'].apply(attend)
```

```python
df
```

```python
df['Math'] = df['Math'].apply(lambda x: min(x, 100))
```
